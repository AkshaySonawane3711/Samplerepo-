#SEABORN

```python
import pandas as pd
import matplotlib.pyplot as mpl
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

UNIVARIATE VARIABLE -- "One variable" Data analysis

```python
tips
```

```python
sns.stripplot(y = 'tip' , data=tips, jitter = True )
plt.ylabel('tips $')
plt.show()
```

```python
sns.stripplot(x= 'day', y= 'tip', data= tips, size= 4)
plt.ylabel('tips ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y= 'tip', data= tips, size = 8)
plt.ylabel('tips')
plt.show()
```

```python
sns.swarmplot(x= 'day', y= 'tip', data= tips, size= 8, hue= 'sex')
plt.ylabel('tips ($)')
plt.show()
```

```python
sns.swarmplot(x= 'tip', y= 'day', data= tips, hue= 'sex', orient= 'h', size=8)
plt.ylabel('tips ($)')
```

```python
sns.boxplot(x= 'day', y= 'tip', data= tips)
plt.ylabel('tips ($)')
```

```python
sns.violinplot(x= 'day', y= 'tip', data= tips)
plt.ylabel('tips ($)')
plt.show()
```

```python
plt.subplot(1,3,1)
sns.violinplot(x= 'day', y= 'tip', data= tips)
plt.ylabel('tips ($)')
plt.subplot(1,3,2)
sns.boxplot(x= 'day', y= 'tip', data= tips)
plt.ylabel('tips ($)')
plt.subplot(1,3,3)
sns.swarmplot(x= 'tip', y= 'day', data= tips, hue= 'sex', orient= 'h', size=8)
plt.ylabel('tips ($)')
```

```python
#combining the plots
sns.violinplot(x= 'day', y= 'tip', data= tips,inner= None, color= "lightgray")
sns.stripplot(x= 'day', y= 'tip', data= tips, hue= 'sex')
```

#Bivarate "two variable" data visualization

```python
sns.jointplot(x= 'total_bill', y= 'tip', data = tips)
plt.show()
```

```python
sns.jointplot(x= 'total_bill', y= 'tip', data = tips,kind='kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips , hue='sex')
plt.show()
```
